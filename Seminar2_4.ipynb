# **Задание1**

1. Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)
2. Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)
3. Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD, SDD и т.д.)
4. Удалите признаки Memory и ScreenResolution

```python
import pandas as pd


df = pd.read_csv('laptop_price.csv')
# df.head()
df.columns
```

```python
# 1
df['Cpu_Company'] = df['Cpu'].apply(lambda x: x.split()[0])
df.tail(10)
```

```python
# 2
def convert_to_gb(x):
  memory = x.split()[0]
  if 'GB' in memory:
    res = memory.replace("GB", "")
  elif "TB" in memory:
    res = float(memory.replace("TB", "")) * 1024
  return int(res)


df['Memory_Amount'] = df["Memory"].apply(convert_to_gb)
df.head()
```

```python
# 3
df['Memory_Type'] = df['Memory'].apply(lambda x: ' '.join(x.split()[1:]))
df.head()
```

```python
# 4
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
df.head()
```
